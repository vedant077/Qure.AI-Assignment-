```python

import pandas as pd

df = pd.read_csv("updated.csv")

gkk = df.groupby(['state_name', 'county_name'])['deaths'].agg('sum').groupby(level=0).nlargest(3)

data = gkk.sort_values(ascending=False)
```

```python
out = pd.DataFrame()
for i in list(df.groupby('state_name')['deaths'].sum().sort_values(ascending=False).head(3).index):
    print(i)
    temp = df[df['state_name']==i]
    a = temp.groupby('county_name')['deaths'].sum().sort_values(ascending=False).head(3)
    out = pd.concat([out,pd.DataFrame({'state_name':i,'county_name':a.index,'Deaths':a.values})],ignore_index=True)
```
